## The Tangent Function

The **tangent** of an angle in a right triangle is defined as:

$$\tan(\theta) = \frac{\text{opposite}}{\text{adjacent}}$$

Given a right triangle where $\theta$ is one of the non-right angles, the *opposite* side is the one across from $\theta$, and the *adjacent* side is the one next to $\theta$ (that isn't the hypotenuse).
